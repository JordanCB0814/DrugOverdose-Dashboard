```python
import pandas as pd

# Assuming you've already grouped your data
# total_deaths_per_year_state is the grouped DataFrame
total_deaths_per_year_state = Clean_Overdose_df.groupby(['Year', 'State Name'])['Data Value'].sum().reset_index()

# Loop through each unique year
for year in total_deaths_per_year_state['Year'].unique():
    # Filter data for the current year
    year_data = total_deaths_per_year_state[total_deaths_per_year_state['Year'] == year]
    
    # Define the file name for each year
    file_name = f'drug_overdose_deaths_{year}.csv'
    
    # Save the filtered data to a CSV file
    year_data.to_csv(file_name, index=False)

    print(f'Saved {file_name}')
```
